```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/Absenteeism_at_work.csv",';')
```

```python
print(df1)
```

```python
df1.drop(['Absenteeism time in hours','Day of the week','Seasons','Disciplinary failure','Son','Reason for absence','Month of absence' ], axis = 1)
```

```python
df1.fillna(method='bfill', axis=1)
```

```python
Q1 = df1['Transportation expense'].quantile(0.25)
Q3 = df1['Transportation expense'].quantile(0.75)
IQR = Q3 - Q1

df1 = df1[~((df1['Transportation expense'] < (Q1 - 1.5 * IQR)) |(df1['Transportation expense'] > (Q3 + 1.5 * IQR)))]

```

```python
Q1 = df1['Body mass index'].quantile(0.25)
Q3 = df1['Body mass index'].quantile(0.75)
IQR = Q3 - Q1

df1 = df1[~((df1['Body mass index'] < (Q1 - 1.5 * IQR)) |(df1['Body mass index'] > (Q3 + 1.5 * IQR)))]
```

```python
corr = df1['Transportation expense'].corr(df1['Body mass index'])
print(round(corr, 2))
```
